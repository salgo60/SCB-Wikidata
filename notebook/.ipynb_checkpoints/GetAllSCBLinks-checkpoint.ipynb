```python
def read_language_codes(file_path): 
    print(f"[DEBUG] Reading language codes from: {file_path}") 
    df = pd.read_csv(file_path) 
    lang_codes = df['Language Code'].dropna().unique().tolist() 
    print(f"[DEBUG] Found {len(lang_codes)} language codes.") 
    return lang_codes 
    
def read_domains(file_path): 
    print(f"[DEBUG] Reading domains from: {file_path}") 
    df = pd.read_csv(file_path, header=None, names=['domain']) 
    domains_list = df['domain'].dropna().unique().tolist() 
    print(f"[DEBUG] Found {len(domains_list)} domains.") 
    return domains_list
```

```python
# -----------------------------------------------------------
# Fetch exturlusage entries for one lang/domain
# -----------------------------------------------------------
def fetch_exturlusage(lang, domain):
    base = f"https://{lang}.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "list": "exturlusage",
        "euquery": domain,
        "eulimit": "max"
    }
    while True:
        r = session.get(base, params=params, timeout=10)
        data = r.json()
        for item in data.get("query", {}).get("exturlusage", []):
            yield {
                "lang": lang,
                "page_title": item.get("title"),
                "url": item.get("url"),
                "wiki_link": f"https://{lang}.wikipedia.org/wiki/{item.get('title').replace(' ', '_')}"
            }

        if "continue" not in data:
            break
        params.update(data["continue"])
        time.sleep(0.3)
```

```python
from pathlib import Path
import pandas as pd

def save_sortable_html(df, out_path):
    out_path = Path(out_path)

    html_table = df.to_html(
        classes="pivot",
        border=0,
        na_rep="0"
    )

    css = """
    <style>
      body {
        font-family: Arial, sans-serif;
        margin: 20px;
      }
      table.pivot {
        border-collapse: collapse;
        width: 100%;
        font-size: 12px;
      }
      table.pivot th, table.pivot td {
        border: 1px solid #999;
        padding: 6px 8px;
        text-align: right;
        white-space: normal;       /* <-- låter header visa hela texten */
      }
      table.pivot th {
        cursor: pointer;            /* <-- visar att header är klickbar */
        background: #f2f2f2;
      }
      table.pivot th:hover {
        background: #e2e2e2;
      }
    </style>
    """

    js = """
    <script>
    // Enkel tabellsortering i ren JS
    document.addEventListener('DOMContentLoaded', () => {
        document.querySelectorAll("table.pivot th").forEach((header, colIndex) => {
            header.addEventListener("click", () => {
                const table = header.closest("table");
                const tbody = table.querySelector("tbody");
                const rows = Array.from(tbody.querySelectorAll("tr"));
                const asc = !header.classList.contains("sorted-asc");

                // Sortera
                rows.sort((a, b) => {
                    const A = a.children[colIndex].innerText.trim();
                    const B = b.children[colIndex].innerText.trim();
                    const numA = parseFloat(A.replace(",", ".")) || 0;
                    const numB = parseFloat(B.replace(",", ".")) || 0;
                    return asc ? numA - numB : numB - numA;
                });

                // Rensa ev tidigare sort-klass
                table.querySelectorAll("th").forEach(th => th.classList.remove("sorted-asc","sorted-desc"));

                // Sätt sorteringsklass för stil
                header.classList.add(asc ? "sorted-asc" : "sorted-desc");

                // Stoppa tillbaka raderna i sorterad ordning
                rows.forEach(row => tbody.appendChild(row));
            });
        });
    });
    </script>
    """

    html = f"""
    <html>
    <head>
    <meta charset="utf-8">
    <title>Sortable Pivot Table</title>
    {css}
    </head>
    <body>
    <h1>Domain-by-language matrix</h1>
    <p>Klicka på valfri kolumnheader för sortering.</p>
    {html_table}
    {js}
    </body>
    </html>
    """

    out_path.write_text(html, encoding="utf-8")
    print("HTML skapad:", out_path)

# Skriv HTML-fil
save_sortable_html(domain_by_lang, "results/domain_by_lang.html")
```

```python

def clean_domain(d):
    """Returnerar en ren domän eller None om ogiltig."""
    if not isinstance(d, str):
        return None

    d = d.strip()
    if d == "":
        return None

    # Om det är en URL, extrahera bara värddelen
    if "://" in d:
        from urllib.parse import urlparse
        d = urlparse(d).netloc

    # Ta bort trailing slash
    d = d.split("/")[0].strip()

    # Måste innehålla en punkt (mycket enkel validering)
    if "." not in d:
        return None

    # Maxlängd per label är 63 i IDNA
    for label in d.split("."):
        if len(label) == 0 or len(label) > 63:
            return None

    return d


import socket, ssl
import pandas as pd
from pathlib import Path


# ============================================================
# A. DOMÄNANALYS (DNS + TLS)
# ============================================================

def domain_dead(domain):
    # 1. DNS-sökning
    try:
        socket.gethostbyname(domain)
    except socket.gaierror:
        return True

    # 2. TLS-anslutning på port 443
    try:
        ctx = ssl.create_default_context()
        with socket.create_connection((domain, 443), timeout=3) as sock:
            ctx.wrap_socket(sock, server_hostname=domain)
        return False
    except:
        return True


def analyze_domains(df):
    domains = sorted([clean_domain(d) for d in df["domain"].unique() if clean_domain(d)])
    dead = []
    alive = []

    for d in domains:
        if domain_dead(d):
            dead.append(d)
        else:
            alive.append(d)

    links_to_dead = df[df["domain"].isin(dead)]

    return {
        "total_links": len(df),
        "total_domains": len(domains),
        "dead_domains": len(dead),
        "alive_domains": len(alive),
        "links_to_dead_domains": len(links_to_dead),
        "dead_domain_list": dead,
        "alive_domain_list": alive,
    }


# ============================================================
# B. HTML-EXPORT AV PIVOTTABELL (SORTERING + FÄRG)
# ============================================================

def pivot_to_html(domain_by_lang):
    # Highlight toppvärden per kolumn
    styled = domain_by_lang.style.apply(
        lambda col: [
            'background-color: #ffdd88' if v == col.max() and col.max() > 0 else ''
            for v in col
        ],
        axis=0
    )

    return styled.to_html(classes="pivot", border=0)


# ============================================================
# C. SCB-RAPPORT SOM ENDA HTML-FIL
# ============================================================

def generate_scb_report(df, domain_by_lang, out_path="scb_lankarapport.html"):
    analysis = analyze_domains(df)
    pivot_html = pivot_to_html(domain_by_lang)

    total_links = analysis["total_links"]
    total_domains = analysis["total_domains"]
    dead_domains = analysis["dead_domains"]
    alive_domains = analysis["alive_domains"]
    links_to_dead_domains = analysis["links_to_dead_domains"]
    dead_list = analysis["dead_domain_list"]
    alive_list = analysis["alive_domain_list"]

    # Sorterbar tabell JS
    js = """
    <script>
    document.addEventListener('DOMContentLoaded', () => {
        document.querySelectorAll("table.pivot th").forEach((header, colIndex) => {
            header.addEventListener("click", () => {
                const table = header.closest("table");
                const tbody = table.querySelector("tbody");
                const rows = Array.from(tbody.querySelectorAll("tr"));
                const asc = !header.classList.contains("sorted-asc");

                rows.sort((a, b) => {
                    const A = a.children[colIndex].innerText.trim();
                    const B = b.children[colIndex].innerText.trim();
                    const numA = parseFloat(A.replace(",", ".")) || 0;
                    const numB = parseFloat(B.replace(",", ".")) || 0;
                    return asc ? numA - numB : numB - numA;
                });

                table.querySelectorAll("th").forEach(th => th.classList.remove("sorted-asc","sorted-desc"));
                header.classList.add(asc ? "sorted-asc" : "sorted-desc");

                rows.forEach(row => tbody.appendChild(row));
            });
        });
    });
    </script>
    """

    css = """
    <style>
    body { font-family: Arial, sans-serif; margin: 20px; }
    h1 { color: #b00; }
    .box { border: 2px solid #b00; padding: 12px; background: #ffecec; margin-bottom: 20px; }
    .ok-box { border: 2px solid #0b6; background: #eaffea; padding: 12px; margin-bottom: 20px; }
    .info-box {
        border: 2px solid #3A7BD5;
        background: #EAF3FF;
        padding: 12px;
        margin-bottom: 20px;
        border-radius: 10px;
    } 
    table.pivot { border-collapse: collapse; width: 100%; font-size: 12px; }
    table.pivot th, table.pivot td { border: 1px solid #999; padding: 6px 8px; text-align: right; }
    .info-box ul {
    margin: 0;
    padding-left: 18px;
    }
    
    .info-box li {
        margin-bottom: 6px;
        line-height: 1.4;
    }
    </style>
    """

    html = f"""
    <!doctype html>
    <html lang="sv">
    <head>
        <meta charset="utf-8">
        <title>SCB – Länkröta i Wikipedia</title>
        {css}
    </head>
    <body>

    <h1>SCB-länkar i Wikipedia – Analys</h1>
    <p><i>Rapport genererad: {generated}</i> / <a href="https://github.com/salgo60/SCB-Wikidata/blob/main/notebook/GetAllSCBLinks.ipynb" target=_blank>Notebook</a></p>
    <div class="info-box">
      <h2>🟢 SCB issues</h2>
      <ul>
        <li><b>SCB-Wikidata:</b> <a href="https://github.com/salgo60/SCB-Wikidata/issues/1" target="_blank">skuggbacklog</a></li>
        <li><b>Issue:</b> <a href="https://github.com/salgo60/SCB-Wikidata/issues/4" target="_blank">#4 Stabila unika identifierare…</a></li>
      </ul>
    </div>

    <div class="ok-box">
    <h2>🟢 Domäner som fortfarande fungerar</h2>
    {", ".join(alive_list)}
    </div>

    <div class="box">
    <h2>🔴 Domäner som är helt döda</h2>
    {", ".join(dead_list)}
    </div>

    <h2>Detaljerad pivottabell (Språk × Domän)</h2>
    {pivot_html}

    {js}

    </body>
    </html>
    """

    #Path(out_path).write_text(html, encoding="utf-8")
    #print("✔ Rapport skapad:", out_path)
    from datetime import datetime

    # 1. Fil med dagens datum
    date_str = datetime.now().strftime("%Y-%m-%d")
    dated_path = out_path.replace(".html", f"_{date_str}.html")
    Path(dated_path).write_text(html, encoding="utf-8")
    
    # 2. Fil utan datum (senaste versionen)
    Path(out_path).write_text(html, encoding="utf-8")
    
    print("✔ Rapport skapad:", dated_path)
    print("✔ Rapport skapad:", out_path)
    
    return out_path


# ============================================================
# D. Kör rapporten
# ============================================================
generate_scb_report(df, domain_by_lang, "results/SCB_lankarapport.html")
```
